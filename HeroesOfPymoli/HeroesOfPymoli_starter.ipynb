```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head(20)
```

```python
#test validity of data
purchase_data_df.isna().mean()
```

```python
#ensure good data, i.e. no nulls
import numpy as np
purchase_data_df.isnull().values.any()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
average_price = purchase_data_df['Price'].mean()
average_price
```

```python
unique_items = len(purchase_data_df['Item Name'].value_counts())
unique_items
```

```python
#total revenue
total_revenue=purchase_data_df["Price"].sum()
total_revenue
```

```python
#combine all values gathered into a new DataFrame
summary_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                              "Average Price": average_price,
                              "Number of Purchases": number_of_purchases,
                              "Total Revenue": total_revenue})
summary_df
```

```python
#format data
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:.2f}".format)
summary_df["Number of Unique Items"] = summary_df["Number of Unique Items"].map("{:,}".format)
summary_df["Number of Purchases"] = summary_df["Number of Purchases"].map("{:,}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
total_count=purchase_data_df.drop_duplicates(subset=["SN"]).groupby(["Gender"]).count()["Purchase ID"]
cum_total_count=total_count.sum()
percentage_of_players=(total_count / cum_total_count) * 100
percentage_of_players
gender_demographics_df = pd.DataFrame({"Total Counts":total_count,"Percentage of Players":percentage_of_players})
gender_demographics_df["Percentage of Players"]=gender_demographics_df["Percentage of Players"].map("{:.2f}%".format)
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count=purchase_data_df.groupby(["Gender"]).count()["Purchase ID"]
average_purchase_price=purchase_data_df.groupby(["Gender"]).mean()["Price"]
total_purchase_value=purchase_data_df.groupby(["Gender"]).sum()["Price"]
total_purchasers=purchase_data_df.drop_duplicates(subset=["SN"]).groupby(["Gender"]).count()["Purchase ID"]
avg_total_purchase_per_person=total_purchase_value / total_purchasers
purchases_by_gender_df = pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":average_purchase_price,
                                      "Total Purchase Value":total_purchase_value,"Avg Total Purchase per Person":avg_total_purchase_per_person})
purchases_by_gender_df["Average Purchase Price"]=purchases_by_gender_df["Average Purchase Price"].map("${:.2f}".format)
purchases_by_gender_df["Total Purchase Value"]=purchases_by_gender_df["Total Purchase Value"].map("${:,.2f}".format)
purchases_by_gender_df["Avg Total Purchase per Person"]=purchases_by_gender_df["Avg Total Purchase per Person"].map("${:.2f}".format)
purchases_by_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
import sys
purchase_data_no_dups_df=purchase_data_df.drop_duplicates(subset=["SN"])
age_demogs_df=pd.DataFrame()
age_series=pd.cut(
    purchase_data_no_dups_df['Age'], 
    bins=[0,9,14, 19, 24, 29, 34,39, sys.maxsize], 
    labels=['<10','10-14','15-19', '20-24', '25-29', '30-34', '35-39', '40+']
)
age_demogs_df["Age Categories"]=age_series
total_count_by_cat=age_demogs_df.groupby(["Age Categories"]).size()
cum_total_count_by_cat=total_count_by_cat.sum()
percentage_of_counts=(total_count_by_cat / cum_total_count_by_cat) * 100
age_demogs_df = pd.DataFrame({"Total Counts":total_count_by_cat,"Percentage of Players":percentage_of_counts})
age_demogs_df["Percentage of Players"]=age_demogs_df["Percentage of Players"].map("{:.2f}%".format)
age_demogs_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchase_df=pd.DataFrame()
age_series=pd.cut(
    purchase_data_df['Age'], 
    bins=[0,9,14, 19, 24, 29, 34,39, sys.maxsize], 
    labels=['<10','10-14','15-19', '20-24', '25-29', '30-34', '35-39', '40+']
)
age_purchase_df["Age Ranges"]=age_series
age_purchase_df["Price"]=purchase_data_df["Price"]
age_purchase_df["Purchase ID"]=purchase_data_df["Purchase ID"]
age_purchase_df["SN"]=purchase_data_df["SN"]
purchase_count1=age_purchase_df.groupby(["Age Ranges"]).size()
average_purchase_price1=age_purchase_df.groupby(["Age Ranges"]).mean()["Price"]
total_purchase_value1=age_purchase_df.groupby(["Age Ranges"]).sum()["Price"]
total_purchasers1=age_purchase_df.drop_duplicates(subset=["SN"]).groupby(["Age Ranges"]).count()["Purchase ID"]
avg_total_purchase_per_person1=total_purchase_value1 / total_purchasers1
purch_by_age_df = pd.DataFrame({"Purchase Count":purchase_count1,"Average Purchase Price":average_purchase_price1,
                                      "Total Purchase Value":total_purchase_value1,"Avg Total Purchase per Person":avg_total_purchase_per_person1})
purch_by_age_df["Average Purchase Price"]=purch_by_age_df["Average Purchase Price"].map("${:.2f}".format)
purch_by_age_df["Total Purchase Value"]=purch_by_age_df["Total Purchase Value"].map("${:,.2f}".format)
purch_by_age_df["Avg Total Purchase per Person"]=purch_by_age_df["Avg Total Purchase per Person"].map("${:.2f}".format)
purch_by_age_df
```
